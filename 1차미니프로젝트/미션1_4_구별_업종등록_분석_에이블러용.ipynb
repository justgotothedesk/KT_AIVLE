#### [실습문제3] 데이터 확인 및 처리

* head 와 tail 을 보았을때, 어느 데이터만 가져와야 할지 생각 해 봅시다.
* 두번째 행부터 데이터가 시작된다.

```python
seoul_business.columns = seoul_business.iloc[0]

seoul_business
```

```python
seoul_business = seoul_business.iloc[1:]
seoul_business
```

```python
# 필요한 열 선택 (자치구, 동, 그리고 종사자 수에 해당하는 열들만 선택)
selected_columns = ['자치구', '동'] + ['한식 일반 음식점업', '두발미용업', '커피전문점', '기타주점업', '일반 교과 학원', '한식 육류요리 전문점']
selected_data = seoul_business[selected_columns]

# '자치구'와 '동' 열을 기준으로 그룹화하고 각 그룹에서 종사자 수를 합산
grouped_data = selected_data.groupby(['자치구', '동']).sum()

# 결과 확인
print(grouped_data)
```

```python
# 중복되는 행 제거
unique_data = grouped_data.groupby(['자치구']).first()

# 필요한 열 선택
selected_columns = ['한식 일반 음식점업', '두발미용업', '커피전문점', '기타주점업', '일반 교과 학원', '한식 육류요리 전문점']

# 필요한 열의 데이터만 선택
result_data = unique_data[selected_columns]

print(result_data)
```

```python
# 4번째 row 삭제
result_data = result_data.drop(result_data.index[-5])
result_data.to_csv('result_data.csv', index=True)

print(result_data)
```

```python
import pandas as pd

# 주어진 데이터프레임
data = {
    '한식 일반 음식점업': ['5', '11', '5', '5', '4', '6', '-', '-'],
    '두발미용업': ['20', '85', '23', '35', '12', '63', '1', '2'],
    '커피전문점': ['4', '6', '-', '-'],
    '기타주점업': ['37', '109', '6', '19', '55', '94', '1', '1', '16', '41'],
    '일반 교과 학원': ['2', '8', '-', '-'],
    '한식 육류요리 전문점': ['-', '-', '34', '102', '16', '92', '8', '32'],
    '합계': ['29,225', '100,183', '18,080', '36,214', '16,255', '57,322', '13,339', '31,371']
}

df = pd.DataFrame(data)

# 쉼표(,)와 하이픈(-) 제거 후 데이터 타입을 int로 변환
for column in df.columns:
    df[column] = pd.to_numeric(df[column].astype(str).str.replace('[,-]', ''), errors='coerce')

# 결과 확인
print(df.dtypes)
print(df)
```
