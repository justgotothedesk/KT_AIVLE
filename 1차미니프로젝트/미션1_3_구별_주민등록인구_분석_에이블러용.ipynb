#### [실습문제3] 데이터 확인 및 처리

* head 와 tail 을 보았을때, 어느 데이터만 가져와야 할지 생각 해 봅시다.
* 데이터가 세번째 줄부터 시작된다
* 서울시의 각 자치구별 남성, 여성 인구 수와 그 합계를 나타내는 데이터프레임 만들기

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
seoul_people = pd.read_csv('/content/drive/MyDrive/KTaivle/1차미니프로젝트/1.3 seoul_people_202401.csv', sep="\t", encoding = "cp949", header = 2)
seoul_people
```

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
# 동별(2),남자,여자,소계.1 데이터만 가져오기
selected_columns = ['동별(2)', '남자', '여자', '소계.1']
seoul_people_selected = seoul_people[selected_columns]
seoul_people_selected
```

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
# 첫 번째 행 제거
seoul_people_selected = seoul_people_selected.drop(0)
seoul_people_selected
```

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
# 동별(2)->자치구, 소계.1->합계로 이름 대체
seoul_people_selected.rename(columns={'동별(2)': '자치구', '소계.1': '합계'}, inplace=True)
seoul_people_selected
```

#### [실습문제4] 데이터 분포 알아보기
* 다양한 변수를 기준으로 그래프를 그려보고 인사이트를 도출해보세요.

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
import seaborn as sns

# 그래프 크기 설정
plt.figure(figsize=(15, 8))

# 남자, 여자, 합계에 대한 그래프
sns.barplot(x='자치구', y='남자', data=seoul_people_selected, label='남자')
sns.barplot(x='자치구', y='여자', data=seoul_people_selected, label='여자')
sns.barplot(x='자치구', y='합계', data=seoul_people_selected, label='합계', alpha=0.7)

# 그래프 제목과 축 이름 설정
plt.title('서울 자치구별 남자, 여자, 합계 인구')
plt.xlabel('자치구')
plt.ylabel('인구 수')

# 범례 표시
plt.legend()

# x축 라벨 회전
plt.xticks(rotation=45, ha='right')

# 그래프 출력
plt.show()

```
