### (1) ëª¨ë¸ë§: LinearRegression

```python
from sklearn.impute import SimpleImputer

# í‰ê· ê°’ìœ¼ë¡œ ê²°ì¸¡ì¹˜ ëŒ€ì²´
imputer = SimpleImputer(strategy='mean')
train_x = imputer.fit_transform(train_x)
test_x = imputer.transform(test_x)

# LinearRegression ëª¨ë¸ í•™ìŠµ
model = LinearRegression()
model.fit(train_x, train_y)
```

```python
# predictë¡œ ì˜ˆì¸¡í•˜ì—¬ y_pred_LRì— ì €ì¥í•´ë³´ì„¸ìš”.

y_pred_LR = model.predict(test_x)
```

### (4) ëª¨ë¸ë§: Self Choice Model

* Self Choice Model: ì•ì„  êµìœ¡ê³¼ì •ì—ì„œ ë°°ìš´ ë¨¸ì‹ ëŸ¬ë‹ ëª¨ë¸ ì¤‘ ì–´ë–¤ ê²ƒì´ë“  ì¢‹ìŠµë‹ˆë‹¤. ì›í•˜ëŠ” ëª¨ë¸ì„ ì„ íƒí•´ì„œ í•™ìŠµì„ ì‹œì¼œë³´ì„¸ìš”.
* Trainê³¼ Testë¡œ ë‚˜ëˆˆ ë°ì´í„°ë¥¼ ê¸°ì¤€ìœ¼ë¡œ Self Choice Modelë¡œ ëª¨ë¸ì„ í•™ìŠµì„ ì§„í–‰í•˜ê³  í‰ê°€ë¥¼ í•´ì£¼ì„¸ìš”.
* ê·¸ë¦¬ê³  ëª¨ë¸ íŒŒì¼ì„ pklë¡œ ì €ì¥í•´ì£¼ì„¸ìš”.
* ì„±ëŠ¥ì§€í‘œ : MSE, R-squared Score(ì†Œìˆ˜ì  5ì§¸ìë¦¬ê¹Œì§€)

```python
# ì•„ë˜ì— í•„ìš”í•œ ì½”ë“œë¥¼ ì‘ì„±í•˜ê³  ê²°ê³¼ë¥¼ í™•ì¸í•©ë‹ˆë‹¤.

from sklearn.ensemble import VotingRegressor
from sklearn.metrics import mean_squared_error, r2_score

models = [
    ('knn', KNeighborsRegressor()),
    ('lr', LinearRegression()),
    ('lasso', Lasso()),
    ('ridge', Ridge()),
    ('elastic_net', ElasticNet()),
    ('decision_tree', DTR()),
    ('random_forest', RFR()),
    ('gradient_boosting', GBR())
]

voting_regressor = VotingRegressor(models)

voting_regressor.fit(train_x, train_y)

y_pred = voting_regressor.predict(test_x)

mse_value = mean_squared_error(test_y, y_pred)
r2_value = r2_score(test_y, y_pred)

print("MSE:", mse_value)
print("R-squared Score:", round(r2_value, 5))
```
